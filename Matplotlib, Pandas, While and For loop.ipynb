## Customizing Charts

```python
# from previous axample
from matplotlib import pyplot as plt
year1 = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]

#to add more data
year = [1960, 1973, 2005] + year1
pop = [2.678, 3.873, 6.4352] + pop1

#to plot the line chart
plt.plot(year, pop) 

#to add X and Y axis labels
plt.xlabel('Year')
plt.ylabel('Population')

#to add chart title
plt.title('World Population Projection')

#to customize the ylabel using yticks
plt.yticks([0, 2, 4, 6, 8, 10], ['o', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

# While Loops

While loop has the syntax:
    while <condition>:
        expression

```python
#examples 
error = 50.0
while error > 1:
    print(error)
    error = error/4
```

```python
number = int(input('pick a random number'))
while number != 0:
    print("connecting.....")
    if number > 0:
        number = number - 1
    else:
        number = number + 1
    print(number) 
```

# For Loops

This uses the syntax:
        for variable in sequence:
             expression

```python
#example
fam = [1.73, 1.68, 1.71, 1.89] 
for height in fam:
    print(height)
```

```python
#to get both the index and the element in the list, use enumerate
for index, height in enumerate(fam):
    print('Index ' + str(index) + ': ' + str(height))
```

```python
#for loops in string 
for c in "Family":
    print(c.capitalize())
```
